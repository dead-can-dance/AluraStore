#1. Análisis de facturación



En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
import pandas as pd

# Lista de URLs
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
]

# Lista vacía para guardar los ingresos
ingresos = []
tiendas = ["tienda","tienda2","tienda3","tienda4"]
# Recorrer las URLs y calcular ingresos
for url in urls:
    df = pd.read_csv(url)
    df["Precio"] = pd.to_numeric(df["Precio"], errors="coerce")  # Asegurar que la columna sea numérica
    ingreso_total = df["Precio"].sum()
    ingresos.append(ingreso_total)  # Agregar el ingreso a la lista

ingresos = [float(ingreso) for ingreso in ingresos]
#asigna el nombre de la tienda a los ingresos
diccionario = dict(zip(tiendas, ingresos))
# Mostrar la lista completa de ingresos
print(f"Ingresos por tienda: {diccionario}")
```

```python
import pandas as pd

def obtener_ingreso_desde_url(url):
    """Lee el CSV desde la URL y calcula el ingreso total."""
    df = pd.read_csv(url)
    df["Precio"] = pd.to_numeric(df["Precio"], errors="coerce")
    return df["Precio"].sum()

def calcular_ingresos(urls):
    """Calcula los ingresos para una lista de URLs."""
    return [float(obtener_ingreso_desde_url(url)) for url in urls]

def asociar_tiendas_con_ingresos(tiendas, ingresos):
    """Asocia nombres de tiendas con sus ingresos en un diccionario."""
    return dict(zip(tiendas, ingresos))

def main():
    urls = [
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
        "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
    ]
    tiendas = ["tienda", "tienda2", "tienda3", "tienda4"]

    ingresos = calcular_ingresos(urls)
    diccionario = asociar_tiendas_con_ingresos(tiendas, ingresos)

    print(f"Ingresos por tienda: {diccionario}")

# Ejecutar solo si es el script principal
if __name__ == "__main__":
    main()
```

# 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.


```python
import pandas as pd

def obtener_categorias_desde_url(url):
  df = pd.read_csv(url)
  categorias = df.groupby('Categoría del Producto')
  return categorias.size().sort_values(ascending=False)

def calcular_categorias(urls):
  categorias = [obtener_categorias_desde_url(url) for url in urls]
  df = pd.DataFrame(categorias)
  df = df.transpose()
  df.columns = ["tienda", "tienda2", "tienda3", "tienda4"]
  return df

calcular_categorias(urls)
```

# 3. Calificación promedio de la tienda


En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.



```python
import statistics as sts

def obtener_promedio_desde_url(url):
  df = pd.read_csv(url)
  promedio = sts.mean(df['Calificación'])
  return promedio

def calcular_promedios(urls):
  promedios = [obtener_promedio_desde_url(url) for url in urls]
  tiendas = ["tienda", "tienda2", "tienda3", "tienda4"]
  promedios = dict(zip(tiendas, promedios))
  return f'el promedio de calificaciones en los productos de los usaurios es de: {promedios}'


calcular_promedios(urls)
```

# 5. Envío promedio por tienda

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.

```python
import statistics as sts

def obtener_costo_promedio_envio_desde_url(url):
  df = pd.read_csv(url)
  promedio = sts.mean(df['Costo de envío'])
  promedio = round(promedio, 2)
  return promedio

def calcular_promedios(urls):
  promedios = [obtener_costo_promedio_envio_desde_url(url) for url in urls]
  tiendas = ["tienda", "tienda2", "tienda3", "tienda4"]
  promedios = dict(zip(tiendas, promedios))
  return f'el promedio de envios por tienda: {promedios}'


calcular_promedios(urls)
```
